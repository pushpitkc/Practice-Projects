#### Outlier removal by using z-score

```python
from scipy.stats import zscore
z = np.abs(zscore(df1))
df2 = df1[(z<3).all(axis=1)]
```

```python
df2.shape
```

```python
df2.skew()
```

#### Outlier removal by using PowerTranformer

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer()
df_transformed = pt.fit_transform(df2)
df_final = pd.DataFrame(df_transformed, columns=df2.columns)
```

```python
df_final.skew()
```

```python
df_final.head(10)
```

## Machine learning Algorithms

```python
y = df_final['Happiness Score']
x = df_final.copy()
x.drop('Happiness Score',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
scdt = sc.fit_transform(x)
data = pd.DataFrame(scdt,columns=x.columns)
data.head(10)
```

Importing necessary libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

```

```python
#Creating train test split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

Models used are LinearRegression,RandomForestRegression, and KNNeighbour Regressor

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred=model.predict(X_test)
Y_pred
```

```python
print("Linear Regression")
print("#################")
print("Mean absolute error:",mean_absolute_error(Y_pred,Y_test))
print("Mean squared error:",np.sqrt(mean_squared_error(Y_pred,Y_test)))
print("R2 score:",r2_score(Y_test,Y_pred))
```

```python
model1 = RandomForestRegressor()
model1.fit(X_train, Y_train)
```

```python
Y_pred=model1.predict(X_test)
Y_pred
```

```python
print("Random Forest Regressor")
print("#######################")
print("Mean absolute error:",mean_absolute_error(Y_pred,Y_test))
print("Mean squared error:",np.sqrt(mean_squared_error(Y_pred,Y_test)))
print("R2 score:",r2_score(Y_test,Y_pred))
```

```python
model2 = KNeighborsRegressor()
model2.fit(X_train, Y_train)
```

```python
Y_pred=model2.predict(X_test)
Y_pred
```

```python
print("KNeighbor Regression Algorithm")
print("##############################")
print("Mean absolute error:",mean_absolute_error(Y_pred,Y_test))
print("Mean squared error:",np.sqrt(mean_squared_error(Y_pred,Y_test)))
print("R2 score:",r2_score(Y_test,Y_pred))
```

From the above algorithms, we can conclude that Random Forest Algorithms has a higher accuracy at predicting the outcome. Let us try to use hyper parameter tuning to improve the performance of the Random Forest Algorithm and see if we can make its accuracy better.

### Hyper-parameter tuning

```python
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 2000, num = 10)]
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
min_samples_split = [2,4,7]
bootstrap = [True, False]
#Creating the random grid
random_grid = {'n_estimators': n_estimators,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'bootstrap': bootstrap}
```

```python
rf = RandomForestRegressor()
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid)                            
rf_random.fit(X_train, Y_train)

rf_random.best_params_
```

```python
Y_pred=rf_random.predict(X_test)
Y_pred
```

```python
print("After Hyper-parameter tuning")
print("KNeighbor Regression Algorithm")
print("##############################")
print("Mean absolute error:",mean_absolute_error(Y_pred,Y_test))
print("Mean squared error:",np.sqrt(mean_squared_error(Y_pred,Y_test)))
print("R2 score:",r2_score(Y_test,Y_pred))
```

Using the hyper-parameter tuning, we are able to better the accuracy of the algorithm.
